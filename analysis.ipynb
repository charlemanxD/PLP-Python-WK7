2. Groupings on a categorical column (for example, species, region, or department) compute the mean of a numerical column for each group.

```python
n_by_region = sales_df.groupby("Region")["Country"].count()
n_by_region
```

Mean of Total revenue by Region

```python
mean_by_category = sales_df.groupby("Region")["Total Revenue"].mean()
mean_by_category
```
